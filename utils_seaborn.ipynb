```python
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.displot(tips['total_bill'], kde=True);
```

```python
sns.jointplot(x='total_bill', y = 'tip', data = tips, kind = 'reg')
```

```python
sns.jointplot(x='total_bill', y = 'tip', data = tips, kind = 'hex')
```

```python
sns.pairplot(tips, hue='sex', palette='viridis')
```

```python
sns.kdeplot(tips['total_bill'])
sns.rugplot(tips['total_bill'])
```

```python
sns.barplot(x='sex', y='total_bill', data=tips)
```

```python
import numpy as np
```

```python
sns.barplot(x='sex', y='total_bill', data=tips, estimator=np.std)
```

```python
sns.countplot(x='sex', data=tips)
```

```python
sns.boxplot(x='day', y='total_bill', data=tips, palette='rainbow')
```

```python
sns.boxplot(x='day', y='total_bill', hue='sex', data=tips, palette='rainbow')
```

```python
sns.violinplot(x='day', y='total_bill', hue='sex', data=tips, palette='rainbow', split=True)
```

```python
sns.swarmplot(x='day', y='total_bill', hue='sex', data=tips, palette='rainbow')
```

## Regression

```python
sns.lmplot(x="total_bill", y = 'tip', data=tips, hue = 'sex')
```

```python
sns.lmplot(x="total_bill", y = 'tip', data=tips, hue = 'sex', markers=['o','v'])
```

```python
sns.lmplot(x="total_bill", y = 'tip', data=tips, hue = 'sex', markers=['o','v'], scatter_kws={'s':100})
```

```python
sns.lmplot(x="total_bill", y = 'tip', data=tips, hue = 'sex', col='time', row='day')
```

```python
sns.lmplot(x="total_bill", y = 'tip', data=tips, hue = 'sex', aspect=1.6, size=5)
```

## Heatmaps

```python
flights = sns.load_dataset('flights')
```

```python
flights.head()
```

```python
df_flights_matrix = flights.pivot_table(index='year', columns='month', values='passengers')
```

```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(figsize=(12,8))
sns.heatmap(df_flights_matrix, ax = ax, annot=True, fmt=".0f")
```

```python
sns.clustermap(df_flights_matrix)
```

## Style

```python
sns.set_style('white')
sns.countplot(x='sex', data=tips)
```

```python
sns.set_style('ticks')
sns.countplot(x='sex', data=tips, palette='deep')
```

```python
sns.set_style('darkgrid')
sns.countplot(x='sex', data=tips, palette='deep')
```

```python
sns.set_style('dark')
sns.countplot(x='sex', data=tips, palette='deep')
```

```python
sns.set_style('whitegrid')
sns.countplot(x='sex', data=tips, palette='deep')
```

```python
fig, ax = plt.subplots(figsize=(6,4))
sns.countplot(x='sex', data=tips, palette='deep', ax=ax)
ax.set_xlabel('my label')
```

```python
sns.set_context('paper', font_scale=4)
sns.countplot(x='sex', data=tips, palette='deep')
```
