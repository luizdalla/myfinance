## Filter stocks

```python
filtro1 = (df.loc[(df['pl'] > 0) & (df['pl'] < 10) & (df['mrgebit'] > 0.4)]).copy()
filtro2 = (df.loc[(df['pl'] > 0) & (df['pl'] < 10) & (df['mrgebit']> 0.4) & (df['dy'] > 0.1)]).copy()
filtro1.sort_values('dy', ascending=False, inplace=True)
filtro2.sort_values('dy', ascending=False, inplace=True)
```

```python
filtro1.sort_values('dy', ascending=False, inplace=True)
```

```python
plt.figure(figsize = (20,6))
plt.bar(filtro1.index, filtro1.dy)
plt.xticks(rotation=45);
plt.title("Acoes com 0<PL<5, margem Ebit >40% e ordenadas por DY");
```

```python
plt.figure(figsize = (20,6))
plt.bar(filtro2.index, filtro2.dy)
plt.xticks(rotation=45);
plt.title("Acoes com PL>0, margem Ebit >40% e ordenadas por DY");
```
