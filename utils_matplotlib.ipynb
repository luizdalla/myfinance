## Basics

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,10)
y = x**2
```

```python
# oop matplotlib
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) #left, bottom, widthm height
axes.plot(x, y)
axes.set_xlabel("Label x")
axes.set_ylabel("Label y")
axes.set_title("Title")
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) #left, bottom, width, height
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3]) #left, bottom, width, height

axes.plot(x, y, c='b')
axes.set_xlabel("Label x")
axes.set_ylabel("Label y")
axes.set_title("Title")

axes2.plot(y, x, c='r')
axes2.set_xlabel("Label x2")
axes2.set_ylabel("Label y2")
axes2.set_title("Title2")
```

## subplots()

```python
fig, ax = plt.subplots()
ax.plot(x, y, 'r--')
ax.set_title('subplot')
```

```python
fig, ax = plt.subplots(1, 2)
for axes in ax:
  axes.plot(x, y, "r")
  axes.set_title('subplot')
```

```python
fig, ax = plt.subplots(1, 2)
ax[0].plot(x,y,'r')
ax[1].plot(y,x,'g')
plt.tight_layout()
```

```python
fig, ax = plt.subplots(1, 2, figsize=(12,8))
ax[0].plot(x,y,'r')
ax[1].plot(y,x,'g')
```

```python
fig, ax = plt.subplots(2,2)
plt.tight_layout()
```

##legend, labels, titles

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x, y, 'b--',label = 'x, y', lw=8, alpha=0.5)
ax.plot(y, x, 'g--',label = 'y, x')
ax.plot(x, 10*x,label = 'x, x', marker='d', c='black')
ax.set_title('title')
ax.set_xlabel('x label')
ax.set_ylabel('y_label')
ax.legend(loc=6)
```

```python
fig, axes = plt.subplots(1,3, figsize=(12, 4))
axes[0].plot(x, x**2)
axes[0].plot(x, x**3)
axes[0].set_title('Default')

axes[1].plot(x, x**2)
axes[1].plot(x, x**3)
axes[1].axis("tight")
axes[1].set_title('Tigth axes')

axes[2].plot(x, x**2)
axes[2].plot(x, x**3)
axes[2].set_title('Default')
axes[2].set_ylim([0,100])
```
